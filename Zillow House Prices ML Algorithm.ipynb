## The Algorithm

Now that I have the data I need, I can work on my model.

```python
# Installing required packages
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from keras.layers import Dense
```

```python
# Plotting prices for each bedroom from 2000-2023

# X will be the same for all plots. 
x = pd.Series(bedroom_1.columns)
del x[0]
x = pd.to_datetime(x)

plt.style.use('seaborn')
fig, ax = plt.subplots(2,2)

# 1 Bedroom
y_1 = pd.Series(bed_1)
ax[0,0].plot(x, y_1)
ax[0,0].set_title("1 Bedroom House Prices in GA", fontsize=10)
ax[0,0].set_xlabel("Year", fontsize=8)
ax[0,0].set_ylabel("House Price", fontsize=8)
ax[0,0].tick_params(axis='both', which='major', labelsize=8)


# 2 Bedroom
y_2 = pd.Series(bed_2)
ax[0,1].plot(x, y_2)
ax[0,1].set_title("2 Bedroom House Prices in GA", fontsize=10)
ax[0,1].set_xlabel("Year", fontsize=8)
ax[0,1].set_ylabel("House Price", fontsize=8)
ax[0,1].tick_params(axis='both', which='major', labelsize=8)


# 3 Bedroom
y_3 = pd.Series(bed_3)
ax[1,0].plot(x, y_3)
ax[1,0].set_title("3 Bedroom House Prices in GA", fontsize=10)
ax[1,0].set_xlabel("Year", fontsize=8)
ax[1,0].set_ylabel("House Price", fontsize=8)
ax[1,0].tick_params(axis='both', which='major', labelsize=8)

# 4 Bedroom
y_4 = pd.Series(bed_4)
ax[1,1].plot(x, y_4)
ax[1,1].set_title("4 Bedroom House Prices in GA", fontsize=10)
ax[1,1].set_xlabel("Year", fontsize=8)
ax[1,1].set_ylabel("House Price", fontsize=8)
ax[1,1].tick_params(axis='both', which='major', labelsize=8)

fig.tight_layout()
plt.show()
```

When building the model, it's important to apply an algorithm that will fit the data. I will test Stochastic Gradient Descent and the Mean Squared Error as my optimizer and loss function. I currently only used the datasets for bedrooms ranging from 1-4. The reason for this is that I will use the dataset for 5 or more bedrooms to determine the accuracy of my model. 

```python
# Plot data for 5+ bedrooms.

plt.style.use('seaborn')
fig, ax = plt.subplots()

y_5 = pd.Series(bed_5)
ax.plot(x, y_5)
ax.set_title("5+ Bedroom House Prices in GA", fontsize=15)
ax.set_xlabel("Year", fontsize=10)
ax.set_ylabel("House Price", fontsize=10)
ax.tick_params(axis='both', which='major', labelsize=8)

plt.show()
```

```python
# I will use the average of each dataset to predict the average house price for 5+ bedrooms. 

bed_1_avg = bed_1.mean()
bed_2_avg = bed_2.mean()
bed_3_avg = bed_3.mean()
bed_4_avg = bed_4.mean()
bed_5_avg = bed_5.mean()
```

```python
# Build Model
def house_model():
    
    x_train = np.array([1, 2, 3, 4])
    y_train = np.array([bed_1_avg, bed_2_avg, bed_3_avg, bed_4_avg])

    model = Sequential([
        Dense(units=1, input_shape=[1])
    ])
    
    model.compile(optimizer='sgd', loss='mean_squared_error')
    model.fit(x_train, y_train, epochs=2000)
    reduce_tracing = True
    
    return model
```

```python
model = house_model()
```

```python
new_x = 5
prediction = model.predict([new_x])
print(prediction)
```
